```python
# Pandas: Data Analysis Toolkit: https://pandas.pydata.org/pandas-docs/stable/10min.html
import pandas as pd
import matplotlib as mpl
import matplotlib.pyplot as plt
# Seaborn: Statistical Visualization Package: https://seaborn.pydata.org/
import seaborn as sns
import numpy as np
import statsmodels.formula.api as sm
```

```python
# keeps visualizations in line
%matplotlib inline
```

```python
# read nutrition data
data = pd.read_csv("../data/nutrition.csv")
```

```python
# view data (48772 rows x 33 columns)
data
```

```python
# generate descriptive statistics
data.describe()
```

```python
# sorting by columns
data.sort_values(by='Question')
```

```python
# create a new data frame for all the data points with 'percent of adults aged 18 years and older who have obesity'
obesity_data = data.loc[data['Question'] == 'Percent of adults aged 18 years and older who have obesity']
```

```python
obesity_data
```

```python
# For the question 'percent of adults aged 18 years and older who have obesity', 
# what is the age stratification vs. obesity rate

age_obesity_data = obesity_data.loc[data['StratificationCategory1'] == 'Age (years)']
ages = age_obesity_data.groupby(['Stratification1'])[["Data_Value"]].mean()['Data_Value'].index.tolist()
obesity_rates_by_age = age_obesity_data.groupby(['Stratification1'])[["Data_Value"]].mean()['Data_Value'].values.tolist()

print "ages and obesity rates are the same length: ", len(ages) == len(obesity_rates_by_age), "\n"
print "ages: ", ages, "\n"
print "obesity rates by age: ", obesity_rates_by_age, "\n"
```

```python
# create a seaborn bar graph
# For the question 'percent of adults aged 18 years and older who have obesity', what is the location vs. obesity rate

sns.set(style="white", context="talk")
rs = np.random.RandomState(7)

# Set up the matplotlib figure
f, (ax1) = plt.subplots(1, 1, figsize=(8, 5), sharex=True)

# Generate some sequential data
sns.barplot(ages, obesity_rates_by_age, palette="BuGn_d", ax=ax1)
ax1.set_ylabel("Obesity Rate")
ax1.set_xlabel("Age Category")

sns.despine(bottom=True)
```

```python
# Linear Regression
X = age_obesity_data.YearStart # start year
Y = age_obesity_data.Data_Value # obesity rate
est = sm.OLS(Y, X).fit() # linear regression
est.summary()
```

```python
# group by and sum data values by location
locations = obesity_data.groupby(['LocationAbbr'])[["Data_Value"]].mean()['Data_Value'].index.tolist()
obesity_rates = obesity_data.groupby(['LocationAbbr'])[["Data_Value"]].mean()['Data_Value'].values.tolist()

print "Locations and Obesity Rates are the same length: ", len(locations) == len(obesity_rates), "\n"
print "Location Abbreviation: ", locations, "\n" # generate x values
print "Obesity Rate: ", obesity_rates, "\n" # generate y values
```

```python
# create a seaborn bar graph
# For the question 'percent of adults aged 18 years and older who have obesity', what is the location vs. obesity rate

sns.set(style="white", context="talk")
rs = np.random.RandomState(7)

# Set up the matplotlib figure
f, (ax1) = plt.subplots(1, 1, figsize=(20, 6), sharex=True)

# Generate some sequential data
sns.barplot(locations, obesity_rates, palette="BuGn_d", ax=ax1)
ax1.set_ylabel("Obesity Rate")
ax1.set_xlabel("Location Abbreviation")

sns.despine(bottom=True)
```
